```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


## Method 1
df = pd.read_csv('income.csv')

col1, col2, col3, col4 = [], [], [], []
for c in df['native-country'].unique():
    this_part_of_df = df[df['native-country']==c]
    
    col1.append(c)
    col2.append(this_part_of_df.shape[0])
    col3.append(this_part_of_df.mean()['hours-per-week'])
    col4.append(100*this_part_of_df[this_part_of_df['gender']=='Female'].shape[0]/this_part_of_df.shape[0])
    
table_to_show = pd.DataFrame(zip(col1, col2, col3, col4), columns=['Country', 'Number of Users', 'Mean hours of work', '% Females'])
table_to_show
```
