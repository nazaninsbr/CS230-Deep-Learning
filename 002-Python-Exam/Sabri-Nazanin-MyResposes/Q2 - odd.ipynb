```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


## alef
df = pd.read_csv('income.csv')

## b
df['workclass'] = df['workclass'].replace({'?':np.nan})
df['occupation'] = df['occupation'].replace({'?':np.nan})

## jim
print('Number of nans in occupation:', df['occupation'].isna().sum(axis = 0))
print('Number of nans in occupation:', df['workclass'].isna().sum(axis = 0))
print('At least one column is nan:', df[df.isna().any(axis=1)].shape[0])

## dal
df_with_nan = df[df.isna().any(axis=1)]
display(df_with_nan.head(2))
print('table of rows with nan was created')


## he
df_max_work = df.sort_values(by='hours-per-week', ascending=False)
display(df_max_work.head(10))


## vav
df_iran = df[df['native-country']=='Iran']
print('Mean hours per week for iranians:', df_iran.mean()['hours-per-week'])

## ze
print('% female iranians:', 100 * df_iran[df_iran['gender']=='Female'].shape[0]/df_iran.shape[0])
print('% male iranians:', 100 * df_iran[df_iran['gender']=='Male'].shape[0]/df_iran.shape[0])
```
