```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


## alef
df = pd.read_csv('income.csv')
```

```python
## b
gender_counts = {}
for val in df.income.unique():
    df_of_income = df[df['income']==val]
    gender_counts[val] = [sum(df_of_income['gender']=='Male'), sum(df_of_income['gender']=='Female')]
    
fig = plt.figure()
ax = fig.add_subplot(111)
N = 2
ind = np.arange(N)                
width = 0.35                      

## the bars
rects1 = ax.bar(ind, gender_counts['<=50K'], width,
                color='blue')

rects2 = ax.bar(ind+width, gender_counts['>50K'], width,
                color='red')

ax.set_ylabel('Number of people')
ax.set_title('Gender Counts based on Income')
xTickMarks = ['Male', 'Female']
ax.set_xticks(ind+width/2)
xtickNames = ax.set_xticklabels(xTickMarks)
plt.setp(xtickNames, rotation=45, fontsize=10)

ax.legend( (rects1[0], rects2[0]), ('<=50K', '>50K') )
plt.show()
```

```python
## jim

marital_status, counts_of_status, proportion_of_status = [], [], []
for val in df['marital-status'].unique():
    df_this_status = df[df['marital-status']==val]
    marital_status.append(val)
    counts_of_status.append(df_this_status.shape[0])
    proportion_of_status.append(df_this_status.shape[0]/df.shape[0])

plt.figure()
plt.figure(figsize=(15, 5))
plt.xlabel('Marital Status')
plt.ylabel('Number of People')
plt.title('Distribution of marital-status')
plt.bar(marital_status, counts_of_status)

plt.figure()
plt.figure(figsize=(15, 5))
plt.xlabel('Marital Status')
plt.ylabel('Proportion of People')
plt.title('Distribution of marital-status')
plt.bar(marital_status, proportion_of_status)

```
