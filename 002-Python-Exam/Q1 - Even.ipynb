```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import jdatetime

def count_number_of_days_to_end_of_1399(day, month, year):
    days = 0
    while True:
        days += 1
        if month < 7:
            if day == 31:
                day = 1
                month += 1
            else:
                day += 1
        elif month < 12:
            if day == 30:
                day = 1
                month += 1
            else:
                day += 1
        elif month == 12:
            if year == 1399:
                if day < 30:
                    day += 1
            else:
                if day < 29:
                    day += 1
        
        if month == 12 and day == 29 and not year == 1399:
            days += 1
            year += 1
            month = 1
            day = 1
                
        if month == 12 and day == 30 and year == 1399:
            break
        
    return days 

def date_difference(s, t):
    s_year, s_month, s_day = [int(val) for val in s.split('/')]
    t_year, t_month, t_day = [int(val) for val in t.split('/')]
    s_fa_date = jdatetime.date(s_year, s_month, s_day, locale='fa_IR')
    t_fa_date = jdatetime.date(t_year, t_month, t_day, locale='fa_IR')
    
    delta = s_fa_date - t_fa_date
    print(delta.days)
    
    s_distance_to_esfand_29_1399 = count_number_of_days_to_end_of_1399(s_day, s_month, s_year)
    t_distance_to_esfand_29_1399 = count_number_of_days_to_end_of_1399(t_day, t_month, t_year)
    
    diff = -1*(s_distance_to_esfand_29_1399-t_distance_to_esfand_29_1399)
    diff = diff
    print(diff)
    print('**************')
    
date_difference('1399/10/12','1399/7/5')
date_difference('1398/10/12','1397/7/5')
date_difference('1398/10/12','1396/11/5')
date_difference('1399/6/12','1396/1/5')
```
