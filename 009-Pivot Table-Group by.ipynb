```python
import pandas as pd
import numpy as np

df = pd.read_csv('002-Python-Exam/income.csv')
df.head()
```

```python
df.rename(columns={'hours-per-week':'hours'}, inplace=True)
```

```python
df.income.unique()
```

```python
df.race.value_counts()
```

```python
df.hours[df['race']=='Black'].mean()
```

```python
pd.pivot_table(df, index=['race'], aggfunc='mean')
```

```python
pd.pivot_table(df, index=['race'], columns=['income'], aggfunc='mean')
```

```python
pd.pivot_table(df, index=['race', 'gender'], columns=['income'], aggfunc='mean').hours
```

```python
df.groupby(by=['gender', 'income']).mean()
```

```python
df.groupby(by=['gender', 'income']).agg(['mean', 'count']).age
```
